```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
txt='Merry alone do it burst me songs Sorry equal charm joy her those folly ham In they no is many both Recommend new contented intention improving bed performed age Improving of so strangers resources instantly happiness at northward Danger nearer length oppose really add now either But ask regret eat branch fat garden Become am he except wishes Past so at door we walk want such sang Feeling colonel get her garrets own'
```

```python
def pos(txt):
    stop_words = set(stopwords.words('english'))
    tokenized = sent_tokenize(txt)
    
    for i in tokenized:
        wordsList = nltk.word_tokenize(i)
        wordsList = [w for w in wordsList if not w in stop_words] 
        tagged = nltk.pos_tag(wordsList)
    return tagged
```
